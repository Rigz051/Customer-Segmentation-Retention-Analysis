```python
import pandas as pd

df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")

df.head()
```

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce").fillna(0)
```

```python

df = df.drop("customerID", axis=1)

df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = le.fit_transform(df[col])

df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in df.columns:
    if df[col].dtype == 'object' and col != 'Churn':
        df[col] = le.fit_transform(df[col].astype(str))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X = df.drop(["Churn", "customerID"], axis=1, errors='ignore')
y = df["Churn"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LogisticRegression(max_iter=10000, solver='saga', tol=0.01)
model.fit(X_train, y_train)

print("Model trained successfully")
```

```python
df['Churn_Prob'] = model.predict_proba(X)[:,1]

def retention_offer(row):
    if row['Churn_Prob'] > 0.7 and row['MonthlyCharges'] > 70:
        return "Premium Offer"
    elif row['Churn_Prob'] > 0.7:
        return "Basic Offer"
    else:
        return "No Offer"

df['Retention_Action'] = df.apply(retention_offer, axis=1)

df.head()
```

```python
df.to_csv("final_output.csv", index=False)

print("Your project is completed!")
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)

print("Model Accuracy:", accuracy)
```
